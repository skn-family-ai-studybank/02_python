# 형변환 (Casting)

- 형변환은 값을 다른 자료형으로 바꾸는 작업이다.
- Python에서는 상황에 따라 자동으로 바뀌는 **암묵적 형변환**과, 개발자가 직접 함수로 바꾸는 **명시적 형변환**이 있다.
- 조건문에서는 값 자체가 `True` 또는 `False`처럼 평가되기도 한다.

## 1. 암묵적 형변환

암묵적 형변환은 Python이 연산 과정에서 자동으로 자료형을 맞춰 주는 경우이다.

- `int + float`는 더 넓은 표현 범위를 가진 `float` 결과가 된다.
- `bool`은 숫자 연산에서 `True == 1`, `False == 0`처럼 처리된다.

```python
print("1 + 2.3 = ", 1 + 2.3) # 1.0(float) + 2.3(float) = 3.3(float)
print('True + 1: ', True + 1) # (int)True == 1
print('True + True: ', True + True) # (int)True == 1
```

## 2. 명시적 형변환

명시적 형변환은 개발자가 `str()`, `int()`, `float()` 같은 함수를 사용해 직접 자료형을 바꾸는 방식이다.

- 문자열과 숫자는 바로 더할 수 없으므로 숫자를 문자열로 바꿔야 한다.
- `int()`로 실수를 정수로 바꾸면 소수점 아래는 버려진다.
- 숫자로 된 문자열은 `float()` 또는 `int()`로 변환해 계산에 사용할 수 있다.

```python
print("'안녕' + 123: ",'안녕'+ str(123)) # 문자열 이어쓰기 / 123 앞에 str로 명시적 형 변환 진행

height = 175.164
print("int(height): ", int(height)) # 소수점 버림
print("float(height): ", float(height))
print("float(123): ", float(123)) # 정수 -> 실수 변환

# parsing : 문자열을 해석해서 타입, 모양을 변경
print("--- 문자열 parsing ---")

str_number = '123456789'
print('파싱 전 : ',str_number ,type(str_number))
print('파싱 후 : ',int(str_number) ,type(int(str_number)))

# print(int('일이삼')) # 파싱 에러
# ValueError: invalid literal for int() with base 10: '일이삼'
```

## 3. 논리형으로의 형변환

조건식에서는 값이 `bool`로 변환되어 평가된다.

- **truthy**: 조건식에서 `True`처럼 평가되는 값
- **falsy**: 조건식에서 `False`처럼 평가되는 값
- 일반적으로 값이 있으면 `True`, 값이 비어 있거나 `0`이면 `False`로 평가된다.
- `''`는 빈 문자열이라 `False`이고, `' '`는 공백 문자 1개가 들어 있으므로 `True`이다.

```python
print("--- bool 형변환 : True ---")
print(bool(1)) # 1을 bool 타입으로 형변환 시 True == 1은 Truthy 하다
print(bool(100111)) # 0을 제외한 모든 정수는 Truthy하다
print(bool(-100))
print(bool(0))
print(bool('abc')) # 문자형은 값만 존재하면 Truthy
print(bool(' '))
print(bool([1,2])) # list, tuple, dict 등 여러 값을 저장하는 자료형에
print(bool((1,2))) # 값이 하나라도 존재하면 Truthy
print(bool({"price": 10000}))
```

```python
print("--- bool 형변환 : False ---")
print(bool(0)) # 0을 bool 타입으로 형변환 시 True == 0은 Falsy 하다
print(bool('')) # 빈 문자열(빈칸)

print(bool([])) # list, tuple, dict 등 여러 값을 저장하는 자료형에
print(bool(())) # 값이 하나라도 존재하지 않으면 Falsy
print(bool({}))

print(bool(None))
```

## 4. 조건문에서의 활용

리스트, 튜플, 딕셔너리 같은 컬렉션은 요소가 있으면 `True`, 비어 있으면 `False`로 평가된다. 그래서 `len(lst) > 0`처럼 직접 길이를 비교하지 않아도 조건문에서 간단히 사용할 수 있다.

```python
lst = [1,2] # Truthy == True

if lst:
    print("lst에 값이 존재합니다")
else:
    print("lst에 값이 존재하지 않습니다")
```

## Getter/Setter 메서드
캡슐화된 클래스에서 속성을 제어하기 위한 메서드로써 기본적으로 다음 두 유형이 있다.
1. getter 속성의 값을 읽어내기 위한 메서드
2. setter 속성의 값을 설정하기 위한 메서드

```python
class MyClass:
    def __init__(self, value):
        self._value = value

    # Getter
    # 외부에서 private 속성 값을 얻어가게 하는 메서드
    def get_value(self):
        return self._value

    # Setter
    # 외부에서 private 속성 값을 변경(세팅)하는 메서드
    def set_value(self, value):
        self._value = value


my_cls = MyClass(100)
print('my_cls.get_value(): ', my_cls.get_value())

print('200 세팅')
my_cls.set_value(200)
print('my_cls.get_value(): ', my_cls.get_value())
```

## Python의 `@property`

Python에서는 Java식 `get_value()`, `set_value()` 메서드도 사용할 수 있지만, 더 Python다운 방식으로는 `@property`를 자주 사용한다.

`@property`를 사용하면 메서드로 값을 검증하면서도, 사용하는 쪽에서는 속성처럼 읽고 쓸 수 있다.


```python
class Score:
    def __init__(self, value: int):
        # 1. 처음 객체를 만들 때 바로 property setter가 호출되도록 합니다.
        self.__value = value

    @property
    def value(self):
        # 2. 실제 값이 저장된 내부 변수(_value)를 반환합니다. (무한루프 탈출!)
        return self.__value

    @value.setter
    def value(self, value):
        if not (0 <= value <= 100):
            raise ValueError('점수는 0~100 사이만 세팅 가능')

        # 3. 검증을 통과한 값을 실제 내부 변수(_value)에 저장합니다.
        self.__value = value


# score = Score(80) # error
score = Score(80)
print('score.value: ', score.value) # 출력: 80

# score.value = 150
```

# 연산자

- 연산자는 값을 계산하거나 비교할 때 사용하는 기호이다.
- Python에서는 산술, 비교, 논리, 멤버십, 식별, 조건 표현식 등 다양한 연산자를 사용한다.
- 여러 연산자가 한 식에 함께 있으면 **연산자 우선순위**에 따라 계산 순서가 정해진다.

## 1. 연산자 우선순위

아래 표는 자주 만나는 연산자 우선순위를 높은 순서에서 낮은 순서로 정리한 것이다.

| 우선순위 | 연산자 | 설명 |
|---:|---|---|
| 1 | `()` | 괄호, 그룹핑 |
| 2 | `**` | 거듭제곱 |
| 3 | `+x`, `-x`, `~x` | 단항 연산자 |
| 4 | `*`, `/`, `//`, `%` | 곱셈, 나눗셈, 정수 나눗셈, 나머지 |
| 5 | `+`, `-` | 덧셈, 뺄셈 |
| 6 | `==`, `!=`, `>`, `<`, `>=`, `<=`, `is`, `in` | 비교, 식별, 멤버십 |
| 7 | `not` | 논리 NOT |
| 8 | `and` | 논리 AND |
| 9 | `or` | 논리 OR |
| 10 | `값1 if 조건 else 값2` | 조건 표현식 |
| 11 | `lambda` | 람다 표현식 |

헷갈릴 때는 괄호를 사용해서 계산 순서를 명확히 표현하는 것이 가장 안전하다.

````python
print('2 + 3 * 4:', 2 + 3 * 4)
print('(2 + 3) * 4:', (2 + 3) * 4)
print('10 > 5 and 3 > 1:', 10 > 5 and 3 > 1)
print('not False or False:', not False or False)
````

## 2. 조건 표현식

조건 표현식은 간단한 `if...else`를 한 줄로 작성하는 문법이다.

```python
참일_때_값 if 조건식 else 거짓일_때_값
```

짧은 값 선택에는 유용하지만, 조건이 복잡해지면 일반 `if...else` 문으로 작성하는 편이 읽기 쉽다.

````python
# print(T인 경우 if 조건 else F인 경우)
age = int(input('Enter your age: '))
print('성인 o'if age >= 20 else '성인 x')
````

## 3. 입력값이 비어 있는지 확인하기

문자열은 값이 있으면 `True`, 빈 문자열이면 `False`로 평가된다. 그래서 이름 입력값이 비어 있는지 조건 표현식으로 간단히 처리할 수 있다.

````python
name = input("이름을 입력하세요") # 2026 06 02
print('이름: '+ name if name else '이름을 입력해주세요')
````
